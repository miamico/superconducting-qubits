# **An Exhaustive Derivation of the Effective Hamiltonian for Transmon Qubits with a Tunable Coupler**

## **The Single Transmon Qubit: An Anharmonic Quantum Oscillator**

The transmon qubit, a cornerstone of modern superconducting quantum computing, is an engineered quantum system designed to optimize the trade-off between coherence and controllability. Its physics is best understood by starting with a classical electrical circuit and progressively introducing the quantum and non-linear elements that define its behavior.

### **From Classical LC Circuit to Quantized Oscillator**

The simplest electrical oscillator is the LC circuit, comprising an inductor $L$ and a capacitor $C$. Its dynamics can be described using the Lagrangian formalism.  We identify the magnetic flux through the inductor, $\Phi$, as the generalized coordinate.  The voltage across the capacitor is $V=\dot{\Phi}$, and the current through the inductor is $I=\Phi/L$.  
The energy stored in the capacitor (kinetic-energy analog) is
$T=\tfrac12 C\dot{\Phi}^{2}$, and the energy in the inductor
(potential-energy analog) is $U=\tfrac{\Phi^{2}}{2L}$.  
Hence
$$
\mathcal L \;=\;T-U \;=\;\tfrac12 C\dot{\Phi}^{2}-\tfrac{\Phi^{2}}{2L}.
$$

To move to a quantum description we first derive the classical Hamiltonian via a Legendre transformation.  The canonical momentum conjugate to $\Phi$ is the charge on the capacitor,
$$
Q \;=\;\frac{\partial\mathcal L}{\partial\dot{\Phi}}
      \;=\;C\dot{\Phi}.
$$

The Hamiltonian is therefore  
$$
H(\Phi,Q)\;=\;Q\dot{\Phi}-\mathcal L
           \;=\;\frac{Q^{2}}{2C}+\frac{\Phi^{2}}{2L}.
$$

This is identical to the simple harmonic-oscillator Hamiltonian
$H=\tfrac{p^{2}}{2m}+\tfrac12 m\omega^{2}x^{2}$, with the analogies
$Q\!\leftrightarrow\! p$ and $\Phi\!\leftrightarrow\!x$.

Canonical quantisation promotes $\Phi\to\hat{\Phi}$ and $Q\to\hat{Q}$ and imposes
$$
[\hat{\Phi},\hat{Q}]=i\hbar.
$$
Hence
$$
\hat{H}_{\mathrm{LC}}
  \;=\;\frac{\hat{Q}^{2}}{2C}+\frac{\hat{\Phi}^{2}}{2L},
$$
whose equally spaced spectrum,
$E_{n}=\hbar\omega\bigl(n+\tfrac12\bigr)$, is unsuitable for a qubit because any drive resonant with $\ket0\!\leftrightarrow\!\ket1$ excites all higher levels.

### **The Josephson Junction as a Non-Linear Inductor**

Replacing the linear inductor with a Josephson junction introduces the required anharmonicity.  A Josephson junction obeys
$$
I = I_{c}\sin\varphi,\qquad
V = \frac{\hbar}{2e}\,\dot{\varphi},
$$
where $\varphi=\tfrac{2\pi}{\Phi_{0}}\Phi$ and $\Phi_{0}=h/2e$.
Integrating the power $P=IV$ gives the potential
$$
U(\varphi) = -E_{J}\cos\varphi,
\qquad
E_{J}=\frac{I_{c}\Phi_{0}}{2\pi}.
$$

### **The Cooper-Pair Box (CPB) and Transmon Hamiltonians**

Shunting the junction with a capacitance $C_{\Sigma}$ yields the CPB:
$$
\hat{H}_{\mathrm{CPB}}
  \;=\;\frac{\hat{Q}^{2}}{2C_{\Sigma}}\;-\;E_{J}\cos\hat{\varphi}.
$$

Define the dimensionless number operator $\hat n=\hat Q/2e$ and an offset charge $n_{g}=C_{g}V_{g}/2e$.  With the charging energy
$E_{C}=e^{2}\bigl/2C_{\Sigma}$,
$$
\hat{H}_{\mathrm{CPB}}
  \;=\;4E_{C}\bigl(\hat n-n_{g}\bigr)^{2}
        -E_{J}\cos\hat{\varphi},\qquad
[\hat{\varphi},\hat n]=i.
$$

### **The Transmon Regime: Engineering Noise Immunity**

Operating in the regime $E_{J}\!\gg\!E_{C}$ (typically $50\!-\!100$) exponentially suppresses charge-noise sensitivity while only algebraically reducing the anharmonicity $\alpha\sim-E_{C}$.

### **Key Parameters: Plasma Frequency and Anharmonicity**

Expanding $ -E_{J}\cos\hat\varphi $ about $\varphi=0$ and discarding the constant:
$$
\hat{H}_{\text{transmon}}
  \;\approx\;
  4E_{C}\hat n^{2}
  +\tfrac12 E_{J}\hat\varphi^{2}
  -\tfrac{E_{J}}{24}\hat\varphi^{4}.
$$

Identifying the quadratic part with a harmonic oscillator gives the plasma frequency
$$
\omega_{p}\;=\;\sqrt{8E_{J}E_{C}}\bigl/\hbar
$$
and, from first-order perturbation theory, the anharmonicity
$$
\alpha \;\approx\; -E_{C}.
$$

Using ladder operators,
$$
\hat\varphi
  = \left(\frac{2E_{C}}{E_{J}}\right)^{\!1/4}\!
    \frac{\hat a+\hat a^{\dagger}}{2},\qquad
\hat n
  = i\left(\frac{E_{J}}{2E_{C}}\right)^{\!1/4}\!
    \frac{\hat a^{\dagger}-\hat a}{2},
$$
so
$$
\hat H \;=\;\hbar\omega_{p}\Bigl(\hat a^{\dagger}\hat a+\tfrac12\Bigr)
           +\alpha\,\hat a^{\dagger}\hat a^{\dagger}\hat a\hat a.
$$


## **The Full Circuit Hamiltonian for Three Capacitively Coupled Transmons**

To understand the interaction between qubits, we now generalise the single-transmon model to a system of three transmons: two data qubits ($Q_{1},Q_{2}$) and a central tunable coupler ($C$).  All three are capacitively coupled to each other.  This section derives the full quantum Hamiltonian for this composite system, which is the starting point for all subsequent analysis.

### **Lagrangian Formulation using Node-Flux Variables**

We label the node fluxes by $\Phi_{1},\Phi_{2},\Phi_{c}$.  The potential energy is the sum of the Josephson energies:

$$
U \;=\; -E_{J1}\cos\varphi_{1}
        -E_{J2}\cos\varphi_{2}
        -E_{Jc}(\Phi_{\text{ext}})\cos\varphi_{c},
$$

with the SQUID tunability

$$
E_{Jc}(\Phi_{\text{ext}})
  \;=\;
  E_{Jc,\max}\Bigl|\cos(\pi\Phi_{\text{ext}}/\Phi_{0})\Bigr|.
$$

The kinetic energy stored in the capacitor network is

$$
T \;=\;
\frac12
\sum_{i,j\in\{1,2,c\}}
C_{ij}\,\dot{\Phi}_{i}\dot{\Phi}_{j}
=
\frac12\,
\dot{\boldsymbol{\Phi}}^{\!\mathsf T}
C\,
\dot{\boldsymbol{\Phi}}.
$$

Here $C$ is the $3\times3$ capacitance matrix whose
diagonal elements are total node capacitances and whose off-diagonal
elements are $-C_{ij}$.

### **Canonical Quantisation and the Multi-Mode Hamiltonian**

The canonical charges are
$$
\mathbf Q
  \;=\;
  \frac{\partial\mathcal L}{\partial\dot{\boldsymbol{\Phi}}}
  \;=\;
  C\,\dot{\boldsymbol{\Phi}},
\qquad
\dot{\boldsymbol{\Phi}}=C^{-1}\mathbf Q.
$$

Hence the classical Hamiltonian is

$$
H
  \;=\;
  \frac12\mathbf Q^{\mathsf T}C^{-1}\mathbf Q
  \;-\;
  \sum_{k=1,2,c}E_{Jk}\cos\varphi_{k}.
$$

After promotion to operators and imposing
$[\,\hat{\Phi}_{i},\hat Q_{j}\,]=i\hbar\delta_{ij}$,
the quantum Hamiltonian reads

$$
\hat H
  \;=\;
  \frac12
  \sum_{i,j}(C^{-1})_{ij}\,
  \hat Q_{i}\hat Q_{j}
  \;-\;
  \sum_{k}E_{Jk}(\Phi_{\text{ext}})\cos\hat\varphi_{k}.
$$

### **Decomposing the Hamiltonian: $\hat H=\hat H^{(0)}+\hat V$**

Define the unperturbed part as the sum of three independent transmons,

$$
\hat H^{(0)}
  \;=\;
  \sum_{k=1,2,c}
    \Bigl[
      \tfrac12(C^{-1})_{kk}\hat Q_{k}^{2}
      -E_{Jk}\cos\hat\varphi_{k}
    \Bigr],
$$

and collect the off-diagonal charging terms in

$$
\hat V
  \;=\;
  \sum_{i\neq j}
    \tfrac12(C^{-1})_{ij}\,
    \hat Q_{i}\hat Q_{j}.
$$

### **The Schrieffer–Wolff Transformation and the Dispersive Regime**

A unitary $\hat U=e^{\hat S}$ is constructed so that  
$\hat H_{\text{eff}} = e^{\hat S}\hat H e^{-\hat S}$ becomes block-diagonal between the low-energy (qubit) subspace and the high-energy (coupler) subspace.  To first order we impose

$$
\hat V + [\hat S,\hat H^{(0)}]=0
\quad\Longrightarrow\quad
S_{mn} = \frac{V_{mn}}{E_{m}-E_{n}}
\ (E_{m}\neq E_{n}),
$$

which is valid provided the couplings $g_{1c},g_{2c}\!\ll\!\Delta_{1c},\Delta_{2c}$ (the **dispersive condition**).

Keeping terms up to second order gives

$$
\hat H_{\text{eff}}
  \;\approx\;
  \hat H^{(0)}
  +\tfrac12[\hat S,\hat V],
$$
whose projection onto the computational basis yields the desired effective two-qubit Hamiltonian.

## **Analysis of the Effective Hamiltonian: Qubit–Qubit Interactions**

### **Effective Transverse (XX) Coupling**

Within the rotating-wave approximation,

$$
\hat V_{\text{RWA}}
  =
  g_{1c}\bigl(\hat a_{1}^{\dagger}\hat a_{c}
              +\hat a_{c}^{\dagger}\hat a_{1}\bigr)
  +
  g_{2c}\bigl(\hat a_{2}^{\dagger}\hat a_{c}
              +\hat a_{c}^{\dagger}\hat a_{2}\bigr).
$$

Choosing

$$
\hat S \;\approx\;
\frac{g_{1c}}{\Delta_{1c}}
  \bigl(\hat a_{1}^{\dagger}\hat a_{c}
        -\hat a_{c}^{\dagger}\hat a_{1}\bigr)
+
\frac{g_{2c}}{\Delta_{2c}}
  \bigl(\hat a_{2}^{\dagger}\hat a_{c}
        -\hat a_{c}^{\dagger}\hat a_{2}\bigr),
$$
($\Delta_{ic}=\omega_{i}-\omega_{c}$),
one finds

$$
\hat H_{12,\text{eff}}
  =
  g_{\text{eff}}
  \bigl(\hat a_{1}^{\dagger}\hat a_{2}
        +\hat a_{2}^{\dagger}\hat a_{1}\bigr),
\qquad
g_{\text{eff}}
  =\frac{g_{1c}g_{2c}}{2}
   \left(
       \frac{1}{\Delta_{1c}}
      +\frac{1}{\Delta_{2c}}
   \right).
$$

### **Longitudinal (ZZ) Interaction**

Because of anharmonicity, the state-dependent AC-Stark shifts yield

$$
\zeta_{ZZ}
  =
  (E_{11}-E_{10})-(E_{01}-E_{00}),
$$

and perturbation theory gives the illustrative form

$$
\zeta_{ZZ}
  \;\approx\;
  g_{\text{eff}}^{2}
  \Biggl[
      \frac{1}{\Delta_{1c}}
     +\frac{1}{\Delta_{2c}}
     -\frac{1}{\Delta_{1c}-\alpha_{c}}
     -\frac{1}{\Delta_{2c}-\alpha_{c}}
  \Biggr]
  \;+\;\cdots,
$$
showing how destructive interference can cancel residual ZZ coupling when frequencies are properly chosen.

### **Tunability of the Couplings**

Flux-biasing the coupler SQUID shifts $\omega_{c}\propto\sqrt{E_{Jc}(\Phi_{\text{ext}})}$, thereby controlling
$\Delta_{1c},\Delta_{2c}$ and hence both $g_{\text{eff}}$ and $\zeta_{ZZ}$:

1. **Idle** — large detuning: $g_{\text{eff}}\!\approx\!0$, $\zeta_{ZZ}\!\approx\!0$.  
2. **Gate** — pulse $\omega_{c}$ near $\omega_{1,2}$ to turn $g_{\text{eff}}$ on for a controlled duration.  
3. **Return** — restore the idle flux bias.

| Interaction | Symbol | Approximate formula | Physical mechanism |
| :- | :- | :- | :- |
| Transverse (XX) | $g_{\text{eff}}$ | $\displaystyle\frac{g_{1c}g_{2c}}{2}\Bigl(\frac{1}{\omega_{1}-\omega_{c}}+\frac{1}{\omega_{2}-\omega_{c}}\Bigr)$ | Virtual-photon exchange via the coupler |
| Longitudinal (ZZ) | $\zeta_{ZZ}$ | See equation above | Interference of anharmonic Stark-shift pathways |

## **A Comparative Survey of Coupling Architectures**

| Coupling scheme | Gate speed | Idle crosstalk ($\zeta_{ZZ}$) | Scalability | Circuit complexity | Key advantage | Key disadvantage |
| :- | :- | :- | :- | :- | :- | :- |
| Direct fixed | High | High (static) | Low | Low | Fabrication simplicity | Always-on ZZ |
| Bus resonator | Medium | Medium | Medium | Medium | Long-range links | Shared-mode crosstalk, Purcell loss |
| **Tunable coupler** | High | Tunable → low | High | High | Large on/off ratio, ZZ cancellation | Extra hardware & control lines |
